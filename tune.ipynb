```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('file.csv')#load data
```

```python
from sklearn.preprocessing import MinMaxScaler
# Scaling Age
scaler = MinMaxScaler()
train_df['Age'] = scaler.fit_transform(train_df[['Age']])
train_df.head()
```

```python
from sklearn.model_selection import train_test_split
# define determinants and consequence
feature_cols=['Age', 'Gender', 'family_history', 'benefits', 'care_options', 'anonymity', 'leave', 'work_interfere','no_employees']
determinants = train_df[feature_cols]
consequence=train_df.treatment

x_train,x_test,y_train,y_test=train_test_split(determinants,consequence,test_size=0.30,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import roc_curve,accuracy_score,recall_score
```

```python
import time
import psutil
titles = ['LogisticRegression','SVM','RandomForest']
classifers = [
    LogisticRegression(random_state=0,penalty='l2'),
    RandomForestClassifier(bootstrap=True, random_state=0),
    LinearSVC(random_state=0, tol=1e-5)
]
for title,classifer in zip(titles,classifers):
    print('\n --- Start Model : %s ----\n'%title)
    start = time.clock()
    md=classifer.fit(x_train,y_train)
    print(md)
    #mem = psutil.virtual_memory()
    #print('ä½¿ç”¨å†…å­˜ï¼š', mem.used)
    y_predict = classifer.predict(x_test)
    end = time.clock()
    score = classifer.score(x_test,y_test)
    recall = recall_score(y_test,y_predict)
    print(score)
    print(recall)
    print('Running time: %s Seconds'%(end-start))
```

```python
from keras import layers
from keras import models
model = models.Sequential()
model.add(layers.Dense(16,activation='relu',input_shape=(9,)))
model.add(layers.Dense(16,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
model.summary()
```

```python
x_train = np.asarray(x_train).astype('float32')
x_test = np.asarray(x_test).astype('float32')
y_train = np.asarray(y_train).astype('float32')
y_test = np.asarray(y_test).astype('float32')
```

```python
import keras.backend as K
def getPrecision(y_true, y_pred):
    TP = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))#TP
    N = (-1)*K.sum(K.round(K.clip(y_true-K.ones_like(y_true), -1, 0)))#N
    TN=K.sum(K.round(K.clip((y_true-K.ones_like(y_true))*(y_pred-K.ones_like(y_pred)), 0, 1)))#TN
    FP=N-TN
    precision = TP / (TP + FP + K.epsilon())#TT/P
    return precision

def getRecall(y_true, y_pred):
    TP = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))#TP
    P=K.sum(K.round(K.clip(y_true, 0, 1)))
    FN = P-TP #FN=P-TP
    recall = TP / (TP + FN + K.epsilon())#TP/(TP+FN)
    return recall
```

```python
model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['accuracy',getRecall])
#model.fit(x_train,y_train,validation_split=0.10,epochs=5,batch_size=8)
model.fit(x_train,y_train,epochs=8,batch_size=8)
```
